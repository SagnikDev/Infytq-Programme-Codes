```python
x=np.arange(0,4*np.pi,0.01)
y_sin=np.sin(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('Sine')
y_cos=np.cos(x)
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('Cosine')
```

```python
topic=['Python','Java','Go','Javascript']
percentage=[200, 90,30, 300]
sizes=percentage
labels=list(topic)
colors=['yellowgreen', 'lightgreen', 'darkgreen', 'gold']
explode=(0.1,0,0.3,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.axis('equal')
plt.legend()
plt.show()
```
